```python
message = 'hi there my name is waleed hesham what is your name'
minKey, maxKey = 16, 512
keys = []
times = []
for key_size in range(minKey, maxKey+1):
    e, d, n = generate_key(key_size)
    print('key_size: ', key_size)
    start = time.time()
    encrypt(message, e, n)
    end = time.time()
    times.append(end - start)
    keys.append([e, d, n])
    
print(times)
```

```python
minKey, maxKey = 16, 50
times = []
for key_size in range(minKey, maxKey+1):
    e, _d, n = generate_key(key_size)

    ciphertext = encrypt(message, e, n)

    start = time.time()

    # I used this built-in function to get the prime factorization instead of creating it using loops
    factors = sympy.factorint(n)
    primes = list(factors.keys())
    p = primes[0]
    q = primes[1]
    phi = (p - 1) * (q - 1)
    d = modinv(e, phi)

    end = time.time()

    times.append(end - start)
    # test the private key
    decrepted_message = decrypt(ciphertext, d, n)
    print(key_size)
    if decrepted_message == message:
        print('Success')
    else:
        print('Fail')
    

print(times)
```
